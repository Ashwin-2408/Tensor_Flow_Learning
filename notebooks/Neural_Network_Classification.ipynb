```python
def plot_decision_boundary(Model,x,y):

  """Plots the decision boundary created by a model"""
  X_min,X_max=x[:,0].min()-0.1,x[:,0].max()+0.1
  y_min,y_max=x[:,1].min()-0.1,x[:,1].max()+0.1
  xv,yv=np.meshgrid(np.linspace(X_min,X_max,100),np.linspace(y_min,y_max,100))
  x_input=np.c_[xv.ravel(),yv.ravel()]
  y_pred=Model.predict(x_input)
  print(y_pred[0])

  if y_pred.ndim > 1 and y_pred.shape[1] > 1:
    y_pred=np.argmax(y_pred,axis=1).reshape(xv.shape)
    print("We are doing MultiClass Classification")
  else:
    print("we are doing binary Classification")
    y_pred=np.round(y_pred).astype(int).reshape(xv.shape)

  plt.contourf(xv, yv, y_pred, levels=[-0.5, 0.5, 1.5], cmap=plt.cm.RdYlBu, alpha=0.5, vmin=0, vmax=1)

  plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k')  # optional: plot original points
  plt.title("Decision Boundary")
  plt.show()

```

```python
plot_decision_boundary(Classification_Model_V2,X,Y)
```

```python
plot_decision_boundary(Classification_Model_v1,X,Y)
```

## Recreating non linear activation functions from scratch

```python
x=tf.range(-10,10,2,dtype=tf.float32)
```

```python
plt.plot(x)
plt.show()
```

```python
#creating out own sigmoid function
def sigmoid(x):
  return 1/(1+tf.exp(-x))
```

```python
x_sigmoid = sigmoid(x)
print(x_sigmoid)
```

```python
plt.plot(x_sigmoid)
plt.show()
```

```python
#Creating the relu function
def relu(x):
  return tf.maximum(0,x)
```

```python
X_relu=relu(x)
plt.plot(X_relu)
plt.show()
```

```python
#tanh activation function
def tanh(x):
  return (tf.exp(x)-tf.exp(-x))/(tf.exp(x)+tf.exp(-x))
```

```python
X_tanh=tanh(x)
plt.plot(X_tanh)
plt.show()
```

### Building a MultiClass Classification Model

```python
X_Train[0].shape
```

```python
X_Train_Label[0].shape
```

```python
MultiClass_Modification_Model = tf.keras.Sequential([tf.keras.Input((28,28)),tf.keras.layers.Flatten(),tf.keras.layers.Dense(10,activation="relu"),tf.keras.layers.Dense(10,activation="relu"),tf.keras.layers.Dense(10,activation="softmax")])
MultiClass_Modification_Model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy,optimizer=tf.keras.optimizers.Adam(),metrics=["accuracy"])
MultiClass_Modification_Model.fit(X_Train,X_Train_Label,epochs=10)
```

```python
MultiClass_Modification_Model.evaluate(X_Test,_X_Test_Label)
```

We can use SparseCategoricalCrossEntropy when the labels are denoted by numbers and we can use CategoricalCrossEntroypy when the labels are one_hot encoded

```python
MultiClass_Modification_Model_v2 = tf.keras.Sequential([tf.keras.Input((28,28)),tf.keras.layers.Flatten(),tf.keras.layers.Dense(10,activation="relu"),tf.keras.layers.Dense(10,activation="relu"),tf.keras.layers.Dense(10,activation="softmax")])
MultiClass_Modification_Model_v2.compile(loss=tf.keras.losses.CategoricalCrossentropy,optimizer=tf.keras.optimizers.Adam(),metrics=["accuracy"])
MultiClass_Modification_Model_v2.fit(X_Train,tf.one_hot(X_Train_Label,depth=10),epochs=10)
```

```python
MultiClass_Modification_Model_v2.evaluate(X_Test,tf.one_hot(_X_Test_Label,depth=10))
```

WE can see that the model has a better accurance when we one hot encoded the data and used the CatgoricalCrossEntrophy comparec to the SparseCategoricalCrossEntrophy

## Finding the Ideal Learning Rate for our MutliClass Classification Model

```python
import tensorflow as tf
import numpy as np
import random

tf.random.set_seed(42)
random.seed(42)
np.random.seed(42)

LR_Multi_Class_Model=tf.keras.Sequential([tf.keras.Input((28,28)),tf.keras.layers.Flatten(),tf.keras.layers.Dense(20,activation="relu"),tf.keras.layers.Dense(20,activation="relu"),tf.keras.layers.Dense(10,activation="softmax")])
LR_Multi_Class_Model.compile(loss=tf.keras.losses.CategoricalCrossentropy,optimizer=tf.keras.optimizers.Adam(),metrics=["accuracy"])
Learning_Rate = tf.keras.callbacks.LearningRateScheduler(lambda epochs:1e-3*10**(epochs/20))
history=LR_Multi_Class_Model.fit(X_Train_Norm,tf.one_hot(X_Train_Label,depth=10),epochs=40,validation_data=(X_Test/255,tf.one_hot(_X_Test_Label,depth=10)),callbacks=[Learning_Rate])
```

```python
import matplotlib.pyplot as plt
import pandas as pd
pd.DataFrame(history.history).plot()
plt.show()
```

```python
lr=1e-3*10**(tf.range(40)/20)
loss=history.history["loss"]
plt.semilogx(lr,loss)
plt.title("Finding Ideal Learning Rate")
plt.xlabel("Learning Rate")
plt.ylabel("loss")
plt.show()
```

The Ideal Learning Rate is 0.01 to 0.001, So lets take 0.005

```python
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

Ideal_Model=tf.keras.Sequential([tf.keras.Input((28,28)),tf.keras.layers.Flatten(),tf.keras.layers.Dense(20,activation="relu"),tf.keras.layers.Dense(20,activation="relu"),tf.keras.layers.Dense(10,activation="softmax")])
Ideal_Model.compile(loss=tf.keras.losses.CategoricalCrossentropy(),optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),metrics=["accuracy"])
Ideal_Model.fit(X_Train_Norm,tf.one_hot(X_Train_Label,depth=10),epochs=20,validation_data=(X_Test/255,tf.one_hot(_X_Test_Label,depth=10)))
```

### Visualizing our Data

```python
def visualize_prediction(model,X,Y,Labels):
  index = random.randint(0,len(X)-1)
  image_input = tf.expand_dims(X[index]/255, axis=0)
  ypred=tf.argmax(model.predict(image_input),axis=1)
  print(ypred.numpy())
  truth_label=Labels[Y[index]]
  predicted_Label = Labels[int(ypred[0].numpy())]
  plt.imshow(X[index],cmap=plt.cm.binary)
  if truth_label==predicted_Label:
    plt.xlabel(f"Predicted:{predicted_Label},Actual_Label:{truth_label}",color="green")
  else:
    plt.xlabel(f"Predicted:{predicted_Label},Actual_Label:{truth_label}",color="red")
```

```python
label_list=[ "T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
visualize_prediction(Ideal_Model,X_Test,_X_Test_Label,label_list)
```
